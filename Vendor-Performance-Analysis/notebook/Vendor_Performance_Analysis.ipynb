```python
vendor_revenue = (
    sales.groupby('VendorName')['SalesDollars']
    .sum()
    .reset_index()
)
```

```python
vendor_cost=(
    purchases.groupby('VendorName')['Dollars']
    .sum()
    .reset_index()
)
```

```python
vendor_profit = vendor_revenue.merge(
    vendor_cost, on = 'VendorName'
)
```

```python
vendor_profit['Profit'] = (
    vendor_profit['SalesDollars'] - vendor_profit['Dollars']
)
```

```python
vendor_revenue=(
    sales.groupby('VendorName')['SalesDollars']
    .sum()
    .reset_index()
    .merge(
        purchases.groupby('VendorName')['Dollars']
    .sum()
    .reset_index(),
    on = 'VendorName'
    ) 
)
vendor_profit['Profit'] = (
    vendor_profit['SalesDollars']
    - vendor_profit['Dollars']
)
```

```python
top10_profit = vendor_profit.sort_values(
    by = 'Profit',
    ascending = False
).head(10)
```

```python
plt.figure(figsize=(10,5))
plt.bar(
    top10_profit['VendorName'],
    top10_profit['Profit']    
)

plt.title('Top 10 Vendors By Profit')
plt.xlabel('Vendor')
plt.ylabel('profit')

plt.xticks(rotation=90)
plt.show()
```

```python
vendor_profit['margin_pct'] = (
    vendor_profit['Profit']/
    vendor_profit['SalesDollars']
)*100
```

```python
plt.figure(figsize=(10,5))
plt.hist(vendor_profit['margin_pct'], bins=20)
plt.title('Profit Margin Distribution')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Number of Vendors')
plt.show()
```

```python
vendor_profit_sorted = vendor_profit.sort_values(
    by = 'Profit',
    ascending = False
)
```

```python
vendor_profit_sorted ['Cumulative_profit'] = (
    vendor_profit_sorted['Profit'].cumsum()
)
```

```python
vendor_profit_sorted['Cumulative_Percent'] = (
    vendor_profit_sorted['Cumulative_profit']
    /vendor_profit_sorted['Profit'].sum()
)*100
```

```python
vendor_profit_sorted['Vendor_Rank'] = range(
    1,
    len(vendor_profit_sorted)+1
)
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    vendor_profit_sorted['Vendor_Rank'],
    vendor_profit_sorted['Cumulative_Percent']
)
plt.axhline(
    y=80,
    linestyle='--'
)

plt.title('Pareto Analysis of Vendor Profit')
plt.xlabel('Vendor_Rank')
plt.ylabel('Cumulative_percentage %')

plt.show()
```

```python
median_margin = vendor_profit['margin_pct'].median()

median_margin
```

```python
vendor_profit['Performance'] = np.where(
    vendor_profit['margin_pct'] > median_margin,
    'Overperformer',
    'Underperformer'
)
```

```python
vendor_profit['Performance'].value_counts()
```

```python
performance_count=(
    vendor_profit['Performance']
    .value_counts()
    .reset_index()
)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    performance_count['Performance'],
    performance_count['count']
)

plt.title('Overperformers vs Underperformers')
plt.xlabel('Performance')
plt.ylabel('Number of Vendors')

plt.show()
```

```python
purchases['PODate'] = pd.to_datetime(purchases['PODate'])
purchases ['ReceivingDate'] = pd.to_datetime(purchases['ReceivingDate'])
purchases['LeadTimeDays'] = (purchases['ReceivingDate'] - purchases['PODate']).dt.days
vendor_lead_time = purchases.groupby('VendorName')['LeadTimeDays'].mean().reset_index().round(2)
print(vendor_lead_time.sort_values(by='LeadTimeDays', ascending = False).head())
```

```python
vendor_summary = vendor_profit.merge(
    vendor_lead_time,
    on='VendorName',
    how='left'
)

vendor_summary.head()
```

## Final Business Story

### Most Important Vendors

The analysis identified Diageo North America Inc, Martignetti Companies, and Pernod Ricard USA as the most important vendors based on revenue generation. These vendors contribute a significant share of total company sales and play a critical role in overall business performance.

### Profit Drivers

Profit analysis revealed that Diageo North America Inc, Martignetti Companies, Pernod Ricard USA, Jim Beam Brands Company, and Bacardi USA Inc generated the highest profits. These vendors not only achieved strong sales volumes but also delivered substantial profitability, making them key strategic partners for the business.

### Vendors Requiring Attention

The analysis identified 13 vendors operating at a loss. Among the weakest performers were Vineyard Brands LLC, Black Cove Beverages, Ira Goldman and Williams, LLP, Highland Wine Merchants LLC, and Adamba Imports Intl Inc. These vendors generated negative profits because purchasing costs exceeded sales revenue. Further investigation is required to evaluate pricing strategies, purchasing decisions, and inventory management practices associated with these suppliers.

### Margin Performance

Vendor profitability varied significantly across the portfolio. Flavor Essence Inc achieved one of the highest profit margins at approximately 98.8%, demonstrating exceptional profitability despite relatively low sales volume. In contrast, Truett Hurst recorded the lowest profit margin at -1478.6%, indicating substantial inventory costs compared to realized sales. This highlights that high revenue alone does not guarantee strong profitability.

### Operational Insights

Lead-time analysis identified Flavor Essence Inc as the slowest supplier, with an average delivery time of approximately 13 days. Longer lead times increase the risk of stock shortages and may require higher safety stock levels or adjusted reorder strategies to ensure product availability.

### Strategic Recommendation

Pareto analysis showed that the top 20% of vendors generated approximately 88.48% of total profit. This demonstrates that company profitability is heavily concentrated among a small group of suppliers. Management should prioritize relationships with these high-performing vendors while continuously monitoring and improving the performance of underperforming suppliers.

### Business Impact

This analysis demonstrates the importance of evaluating vendors using multiple performance metrics, including revenue, profit, profit margin, and lead time. By strengthening relationships with top-performing vendors, addressing loss-making suppliers, and improving inventory planning for slow-delivery vendors, the company can improve profitability and make more informed supplier management decisions.
